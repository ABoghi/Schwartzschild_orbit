```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Helvetica"
})
import warnings
warnings.filterwarnings("ignore")
```

```python
Omega_d = np.linspace(0,2*np.pi,100)
```

```python
def define_case(file_name):
    
    df = pd.read_csv(file_name)
    df = df.dropna()
    df = df.reset_index(drop=True)
    
    r = df.r.values
    x = df.x.values
    y = df.y.values
    try:
        tau = df.tau.values
        t = df.t.values
    except:
        tau = 0 * r
        t = tau
    Omega = df[' "Omega"'].values
    
    return x,y,Omega,r,tau,t
    
```
